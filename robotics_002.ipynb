```python
data = pd.read_csv("./sensor_sample/sensor_data_600.txt",delimiter=" ",header=None,names=("data","time","ir","lidar"))
```

## マルチモーダルなグラフ

```python
data["lidar"].hist(bins=max(data["lidar"])-min(data["lidar"]),align="left")
```

## 横軸が取得順にセンサ値つけた番号、縦軸がセンサ値

```python
data.lidar.plot()
```

```python
data["hour"] = [e//10000 for e in data.time]
d = data.groupby("hour")
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist()
d.lidar.get_group(14).hist()
```

## 時間帯を辞書のキーにして時間ごとの頻度のデータフレームを作成

```python
each_hour = { i : d.lidar.get_group(i).value_counts().sort_index() for i in range(24)}#時間ごとにデータフレームを作成
freqs = pd.concat(each_hour,axis=1)
freqs = freqs.fillna(0)
probs = freqs/len(data)
probs
```

```python
import seaborn as sns
```

## 横軸がhour,縦軸が値

```python
sns.heatmap(probs)
```
